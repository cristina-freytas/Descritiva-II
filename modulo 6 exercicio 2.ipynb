####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# importando 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline      


# carregando arquivo e construindo DataFrame
arquivo = pd.read_csv('sinasci_RO_2019.csv')
df_sinasci = pd.DataFrame(arquivo)
```

```python
# boxsplot da idade da mãe
plt.figure(figsize=(6, 4))  
sns.boxplot(data=df_sinasci, y='IDADEMAE')  
plt.title('Idade da Mãe', color='red')  
plt.ylabel('Idade da Mãe', color='red')  
plt.xlabel('')  

```

####  2) Faça um boxplot do peso do bebê

```python
# boxsplot do peso do bebê
plt.figure(figsize=(6, 4))  
sns.boxplot(data=df_sinasci, y='PESO')  
plt.title('Peso do bebê', color='red')  
plt.ylabel('Peso do bebê em gramas', color='red')  
plt.xlabel('')  
```

####  3) Faça um histograma da idade da mãe

```python
# informações sobre os dados
idade_mae = df_sinasci['IDADEMAE'].unique()
print(idade_mae)
```

```python
# histograma da idade da mãe
plot = sns.displot(data = df_sinasci 
            ,alpha = .25
            ,x = 'IDADEMAE'
            ,bins = 20
            ,element = 'step'
            ,kde = True
           )

plot.set(title='Frequência da Idade da Mãe')
plt.title('Frequência da Idade da Mãe', color='red')

plot.set(xlabel='Idade da Mãe')
ax = plt.gca()
ax.xaxis.get_label().set_color('red')

plot.set(ylabel='Frequência')
ay = plt.gca()
ay.yaxis.get_label().set_color('red')
```
